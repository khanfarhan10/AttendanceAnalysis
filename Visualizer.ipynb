```python
#to do get all date time fields
to_replace=df.columns[2:9]
to_replace
```

```python
days=[]
for i in range (len(to_replace)):
    days.append("Day"+str(i))
days
```

```python
def print_dict(new_dict):
    print ("Dictionary is :  ")
    print("keys: values")
    for i in new_dict:
        print(i, " :", new_dict[i])

def merge_list_to_dict(test_keys,test_values):
    # using dictionary comprehension
    # to convert lists to dictionary
    merged_dict = {test_keys[i]: test_values[i] for i in range(len(test_keys))}
    return merged_dict
REP=merge_list_to_dict(to_replace,days)
print_dict(REP)
```

```python
df.rename(columns = REP, inplace = True) 
df.head(10)
```

```python
# making new data frame with dropped NA values in colums
df = df.dropna(axis = 1, how ='any') 
df.head(10)
```

```python
#now define a function that expects 10 students

def plot_data(filter_df):
    n=filter_df.head(10)
    n['Index'] = list(range(n.shape[0]))
    n.head(15)


    #we will plot the samples for each student
    plt.rcParams.update({'font.size': 22})
    fig = plt.figure(figsize=(20,10))
    ax = plt.subplot(111)
    for index, row in n.iterrows():
        #print(row['c1'], row['c2'])
        Attendance=row[2:-3]
        Attendance=row["Index"]+np.array(Attendance)
        Name=row["Name_Main"]
        ax.plot(days, Attendance,label=Name, linewidth=10)
        #ax.plot(days, Mean_F,label='Females')


    chartBox = ax.get_position()
    ax.set_position([chartBox.x0, chartBox.y0, chartBox.width*0.9, chartBox.height])
    plt.title('Student Attendance for each day')
    handles, labels = ax.get_legend_handles_labels()
    #ax.legend(reversed(handles), reversed(labels), title='Line', loc='upper left')
    ax.legend(reversed(handles), reversed(labels),loc='upper center', bbox_to_anchor=(1.22, 1.02), 
              shadow=True, ncol=1,prop={'size': 33},title='Student Names')
    #ax = plt.gca()
    ax.grid(True)
    ax.set_ylabel('Students')
    #ax.set_xlabel('Days(0-5)')

    # Turn off tick labels
    #alternate absent present absent present
    Y_tick=[]
    Y_tick_labels=[]
    for i in range(10+2):
        Y_tick.append(i)
        if(i%2==0):
            Y_tick_labels.append("A")
        else:
            Y_tick_labels.append("P")
    ax.set_yticks(Y_tick) 
    ax.set_yticklabels(Y_tick_labels)
    #plt.savefig('books_read.png')


    #ax.set_yticks([0, 0.5, 1]) 

    #ax.set_yticklabels(("Val-1", "Val-2", "Val-3")) 
    ##ax.set_yticklabels([])
    #ax.set_xticklabels([])

    plt.show()
```

```python
filter_df=df[df["Total"]>2]
filter_df=filter_df[filter_df["Total"]<5]
plot_data(filter_df)
```
