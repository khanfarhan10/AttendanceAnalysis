```python
import numpy as np
```

```python
def count(df):
    df['Absent'] = np.sum(df == 'A')
    df['Present'] = np.sum(df == 'P')
    return df

df_new=df.apply(count, axis = 1)

df_new
```

```python
# filtering with query method 
query1=df_new.query('Absent == 6') 
```

```python
Absent6days=list(query1["Roll"])
print(Absent6days)
```

```python
Absent6daysName=list(query1["Name_Main"])
print(Absent6daysName)
```

```python
# using dictionary comprehension 
# to convert lists to dictionary 
def merge_list_to_dict(test_keys,test_values):
    merged_dict = {test_keys[i]: test_values[i] for i in range(len(test_keys))} 
    return merged_dict
Dict6DaysAbsent=merge_list_to_dict(Absent6days,Absent6daysName)
print(Dict6DaysAbsent)
```

```python
def print_dict(new_dict):
    #print("keys: values") 
    for i in new_dict: 
        print(i, " :", new_dict[i]) 
print_dict(Dict6DaysAbsent)
```

```python
df_ch=df_new.copy(deep=True)
```

```python
counter=0
big_count=0
for each in headers:
    big_count+=1
    if(type(each)==datetime.datetime):
        counter+=1
        if(counter==1):
            starter=big_count-1
print("Counter :",counter)
print("Starter :",starter)
```

```python
no_of_dates=counter
new_count=0
for i in range(starter,no_of_dates+starter):
    date_string="day_{}".format(new_count)
    df_ch.rename(columns = {headers[i]:date_string}, inplace = True)
    new_count+=1
```

```python
def show_details(main_query):
    Actiondays=list(main_query["Roll"])
    ActiondaysName=list(main_query["Name_Main"])
    DictActiondays=merge_list_to_dict(Actiondays,ActiondaysName)
    print_dict(DictActiondays)
show_details(query6)
```

```python
df_ch.replace(to_replace ="A", value =0,inplace=True) 
df_ch.replace(to_replace ="P", value =1,inplace=True) 
```

```python
df_ch
```

```python
new_headers=list(df_ch.columns.values)
X=df_ch[new_headers[2:6]]
X
```

```python
X.shape
```

```python
# now we want to create our response vector
y = df_ch.day_4
y.shape
```

```python
y
```

```python
# 1. import
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# 2. instantiate model
logreg = LogisticRegression()
# 3. fit 
logreg.fit(X_train, y_train)

```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
